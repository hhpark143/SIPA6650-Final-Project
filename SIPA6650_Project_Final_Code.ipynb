Import CSV:

```python
import pandas as pd

file_path = "https://raw.githubusercontent.com/hhpark143/SIPA6650-Final-Project/main/rDMC_wide_v1.csv"
rDMC = pd.read_csv(file_path)
print(rDMC.head())
```

Create dataframes for US and PRC capabilities in 2004

```python
USA2004_df = rDMC[(rDMC['country'] == 'us') & (rDMC['year'] == 2004)].copy().reset_index(drop=True)
```

```python
PRC2004_df = rDMC[(rDMC['country'] == 'china') & (rDMC['year'] == 2004)]
```

```python
us_airpower_2004 = ['air defence_air to air missiles',
       'air defence_surface to air artillery',
       'air defence_surface to air missiles', 'aircraft_aew',
       'aircraft_attack', 'aircraft_bomber', 'aircraft_comm', 'aircraft_ew',
       'aircraft_fighter', 'aircraft_maritime patrol', 'aircraft_paramilitary',
       'aircraft_recon-surveil', 'aircraft_tanker', 'aircraft_transport',
       'aircraft_utility', 'helicopters_attack', 'helicopters_ew', 'helicopters_maritime',
       'helicopters_observation', 'helicopters_search and rescue',
       'helicopters_transport', 'helicopters_utility']
USA2004_df['USA_airpower_2004'] = USA2004_df[us_airpower_2004].sum(axis=1)

us_seapower_2004 = ['amphibious_landing craft',
       'amphibious_landing ship', 'amphibious_other', 'mine warfare_mine countermeasures',
       'mine warfare_mine layers','patrol and coastal combatants_coastal',
       'patrol and coastal combatants_corvette',
       'patrol and coastal combatants_offshore',
       'patrol and coastal combatants_other',
       'principal surface combatants_aircraft carrier',
       'principal surface combatants_cruisers',
       'principal surface combatants_destroyers',
       'principal surface combatants_frigates', 'submarines_attack', 'submarines_ballistic', 'submarines_coastal',
       'submarines_other', 'unmanned aerial vehicles_combat']
USA2004_df['USA_seapower_2004'] = USA2004_df[us_seapower_2004].sum(axis=1)

us_landpower_2004 = ['anti-tank/anti-infrastructure_artillery',
       'anti-tank/anti-infrastructure_missiles',
       'armoured fighting vehicles_attack',
       'armoured fighting vehicles_engineering and maintenance',
       'armoured fighting vehicles_transport']
USA2004_df['USA_landpower_2004'] = USA2004_df[us_landpower_2004].sum(axis=1)

us_ballistic_2004 = ['ballistic missiles_intercontinental',
       'ballistic missiles_medium-range', 'ballistic missiles_short-range']
USA2004_df['USA_ballistic_2004'] = USA2004_df[us_ballistic_2004].sum(axis=1)

us_landsea_defense_2004 = ['land/sea defence_air to surface ordnance',
       'land/sea defence_surface to surface artillery',
       'land/sea defence_surface to surface missiles']
USA2004_df['USA_landseadefense_2004'] = USA2004_df[us_landsea_defense_2004].sum(axis=1)

us_logistics_2004 = ['logistics and support_command and support',
       'logistics and support_logistics', 'logistics and support_maintenance',
       'logistics and support_replenishment', 'logistics and support_research',
       'logistics and support_special purpose',
       'logistics and support_surveillance and intelligence',
       'logistics and support_training', 'logistics and support_transport',
       'logistics and support_yard craft']
USA2004_df['USA_logistics_2004'] = USA2004_df[us_logistics_2004].sum(axis=1)

us_radar_2004 = ['radars_air defense/early warning', 'radars_communications',
       'radars_environmental', 'radars_fire control',
       'radars_ISR/battlefield surveil', 'radars_navigation']
USA2004_df['USA_radar_2004'] = USA2004_df[us_radar_2004].sum(axis=1)

us_uav_2004 = ['unmanned aerial vehicles_recon-surveil', 'unmanned aerial vehicles_transport']
USA2004_df['USA_uav_2004'] = USA2004_df[us_uav_2004].sum(axis=1)
```
